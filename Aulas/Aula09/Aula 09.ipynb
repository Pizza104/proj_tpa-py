## Introdução a Tabelas Hash

> As Tabelas Hash são estruturas de dados comum para a localização de grandes quantidades de dados. Devido sua eficiência na busca, inserção e remoção de elementos utiliza a indexação dos dados como nos dicionarios ja vistos em aula.

Aplicações: 
   1. Banco de Dados
   2. Cache de memória
   3. Compiladores
   4. Algoritmos de buscas
   5. Big data


```python
class HashTable:
    # Construtor da tabela Hash
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(self.size)]

    # Método que calcula o índice da tabela através da chave
    def hash(self, key):
        return hash(key) % self.size

    # Método que insere um valor através da sua chave 
    def insert(self, key, value):
        index = self.hash(key)
        for y in self.table[index]:
            if y[0] == key:
                y[1] = value
                return
        self.table[index].append([key, value])

    # Método que remove valor através da sua chave
    def remove(self, key):
        index = self.hash(key)
        for i, y in enumerate(self.table[index]):
            if y[0] == key:
                del self.table[index][i]
                return True
        return False

    # Método de busca do valor através da sua chave           
    def get(self, key):
        index = self.hash(key)
        for y in self.table[index]:
            if y[0] == key:
                return y[1]
        return None

if __name__ == "__main__":
    biblioteca = HashTable()
    biblioteca.insert("A", "Alice no País das Maravilhas")
    biblioteca.insert("M", "Marcelino Pão e Vinho")
    biblioteca.insert("O", "O Código Da Vinci")

    print(biblioteca.get("A"))  # Alice no País das Maravilhas
    print(biblioteca.get("B"))  # None

    biblioteca.remove("A")
    print(biblioteca.get("A"))  # None
```
